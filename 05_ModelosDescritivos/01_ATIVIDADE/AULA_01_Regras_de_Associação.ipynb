Neste hands-on, voc√™ aprender√° como descobrir regras de associa√ß√£o a partir de um conjunto de dados de compras usando o algoritmo Apriori. Usaremos a biblioteca mlxtend para implementar o algoritmo e encontrar padr√µes interessantes nas compras.

## Prepara√ß√£o do Ambiente üõ†Ô∏è

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

## Importa√ß√£o e Explora√ß√£o dos Dados üì•üîç

O dataset cont√©m compras feitas, onde cada linha √© uma lista compra

```python
data = pd.read_csv('https://raw.githubusercontent.com/devjaynemorais/modelos_descritivos_curso/main/Aula%2001/dados/store_data.csv', header=None)
```

O Algoritmo de apriori recebe  "transa√ß√µes' como entrada, para transformarmos nosso dataframe em transa√ß√µes precisamos convertelo em uma lista de listas.

```python
lists = data.values.tolist()
```

```python
for column in data.columns:
  arr = data[column].tolist()

  if any([text.startswith(' ') for text in arr if text is not np.nan]):
    display(data[column].unique())
```

Convertemos, por√©m as listas contem muitos nan, precisamos removelos.


```python
cleaned_lists = []

for buy_list in lists:
  result = list(filter(lambda x: x is not np.nan, buy_list))

  ## Removendo espa√ßos extras
  result = [text.strip() for text in result]

  if result:
    cleaned_lists.append(result)
```

### Criando as "transactions" üîçüõí

A fun√ß√£o Apriori espera que os dados estejam em um DataFrame do pandas codificado com "one-hot". Para criar, podemos utilizar a classe TransactionEncoder.



```python
te = TransactionEncoder()
```

```python
transactions = te.fit(cleaned_lists).transform(cleaned_lists)
```

```python
transactions
```

```python
df = pd.DataFrame(transactions, columns=te.columns_)
df.head(3)
```

### Aplicando Apriori

Um conjunto de itens √© considerado "frequente" se ele atender a um limite de suporte especificado pelo usu√°rio. Por exemplo, se o limite de suporte for definido como 0,5 (50%), um conjunto de itens frequente √© definido como um conjunto de itens que ocorrem juntos em pelo menos 50% de todas as transa√ß√µes no banco de dados.

Nesse caso, um conjunto de itens deve aparecer em pelo menos 0.45% das transa√ß√µes para ser considerado frequente.

Uma forma de descobrir esse valor √© encontrando o valor de suporte para cada item:

support(item) = (number of transactions containing item) / (total number of transactions)

```python
## Exemplo

result = df.sum() / len(df.index)

result
```

```python
result.mean()
```

```python
df.shape
```

```python
frequent_itemsets = apriori(df, min_support=0.0045, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=3.0)
```
