```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial import distance
from sklearn.metrics.pairwise import haversine_distances
from Levenshtein import distance
from sklearn.metrics.pairwise import haversine_distances




# Solicitando a palavra ao usuário
palavra_busca = input("Digite a palavra que deseja buscar na letra: ")

# Criando um vetorizador e transformando as letras em uma matriz de contagem de palavras
vectorizer = CountVectorizer()
letras_matrix = vectorizer.fit_transform(df['letra'])


# Transformando a palavra de busca em uma matriz de contagem
palavra_matrix = vectorizer.transform([palavra_busca])

# Calculando a similaridade de cosseno entre a palavra de busca e todas as letras
similaridades = cosine_similarity(letras_matrix, palavra_matrix)
dist = haversine_distances(letras_matrix, palavra_matrix)





# Encontrando a linha com a maior similaridade
linha_mais_similar = df.iloc[similaridades.argmax()]

# Exibindo os resultados
print("\nMúsica mais similar à palavra '{}':".format(palavra_busca))
print("Cantor: ", linha_mais_similar['cantorNome'])
print("Música: ", linha_mais_similar['musicaNome'])
print("Cosseno: ", similaridades.max())
print("Haverside: ", dist.max())
print("Letra:\n", linha_mais_similar['letra'])
```
