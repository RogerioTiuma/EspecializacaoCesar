## Criação de Subplots

````python
# Criar um array de valores x
x = np.linspace(0, 2 * np.pi, 100)
````

````python
# Criar figuras e subplots em uma grade 2x2
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plotar em cada subplot
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('Sen(x)')

axs[0, 1].plot(x, np.cos(x), 'tab:orange')
axs[0, 1].set_title('Cos(x)')

axs[1, 0].plot(x, np.tan(x), 'tab:green')
axs[1, 0].set_title('Tan(x)')

axs[1, 1].plot(x, np.exp(x), 'tab:red')
axs[1, 1].set_title('Exp(x)')

# Ajustar layout para evitar sobreposição de rótulos
plt.tight_layout()
plt.show()
````

## Integração do Pandas com Matplotlib

Podemos definir o tipo de gráfico que desejamos trabalhar de duas formas. A primeira delas é passando o argumento kind dentro da função plot(), assim:

```df.plot(x, y, kind='bar')```

Outra forma de ser feito é utilizando a função específica para o gráfico desejado. Nesse exemplo em específico, seria:

 ```plt.bar(x,y)```

 ### Algumas dicas

- ```bar``` ou ```barh``` para bar plots

- ```hist``` para histograma

- ```box``` para boxplot

- ```kde``` ou ```density``` para density plots

````python
import seaborn as sns
````

````python
titanic = pd.read_csv('https://raw.githubusercontent.com/agconti/kaggle-titanic/master/data/train.csv')
````

````python
titanic.drop(['PassengerId', 'Name', 'Cabin', 'Ticket'],axis=1,inplace=True)
````

````python
# usando seaborn para ver quais colunas possuem valores nulos
titanic.head(5)
````

````python
titanic.info()
````

````python
nvp = [ "#3f52a6", "#CD5C5C"]
````

````python
sns.histplot(x='Age', data=titanic, kde=True)
````

````python
sns.violinplot(y=titanic["Age"], x=titanic["Sex"], linewidth=5)
plt.xlabel("Sexo")
plt.ylabel("Idade")
plt.title("Gráfico do Violino da Idade dos passageiros do Titanic")
````

````python
ax = sns.violinplot(y=titanic["Age"], x=titanic["Sex"], palette=nvp)

# Adicionando título ao gráfico
ax.set_title('Distribuição da Idade por Gênero')

# Renomeando os rótulos dos eixos x e y
ax.set_xlabel('Gênero')
ax.set_ylabel('Idade')

# Adicionando a legenda
# plt.legend(labels=['Mulheres', 'Homens'])

plt.show()
````

````python
cores = sns.color_palette('Blues')
````

````python
cores
````

````python
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Escolhendo uma paleta de cores do Seaborn
colors = sns.color_palette("husl", 2)  # 'husl' é o nome da paleta de cores

# Plotagem do gráfico com personalização de cores a partir da paleta
plt.plot(x, y, color=cores[2], linestyle='--', linewidth=2, label='Seno')  # Usando a primeira cor da paleta
plt.scatter(x, y, color=cores[5], label='Pontos')  # Usando a segunda cor da paleta

# Adicionando título e rótulos dos eixos
plt.title('Exemplo de Personalização de Cores a partir de uma Paleta')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# Adicionando legenda
plt.legend()

plt.show()
````

## Criação de Subplots

````python
# Carregue o conjunto de dados
data = sns.load_dataset("iris")

# Crie uma figura com subplots usando o Matplotlib
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 2 linhas, 2 colunas de subplots

# Crie gráficos individuais para cada subplot usando o Seaborn
sns.scatterplot(x="sepal_length", y="sepal_width", data=data, ax=axes[0, 0])
axes[0, 0].set_title("Comprimento vs Largura da Sépala")

sns.scatterplot(x="petal_length", y="petal_width", data=data, ax=axes[0, 1])
axes[0, 1].set_title("Comprimento vs Largura da Pétala")

sns.boxplot(x="species", y="sepal_length", data=data, ax=axes[1, 0])
axes[1, 0].set_title("Comprimento da Sépala por Espécie")

#sns.boxplot(x="species", y="petal_length", data=data, ax=axes[1, 1])
axes[1, 1].boxplot(data.groupby('species')['petal_length'].apply(list), labels=data['species'].unique())

axes[1, 1].set_title("Comprimento da Pétala por Espécie")

# Ajuste o layout dos subplots
plt.tight_layout()

# Exiba os subplots
plt.show()
````
