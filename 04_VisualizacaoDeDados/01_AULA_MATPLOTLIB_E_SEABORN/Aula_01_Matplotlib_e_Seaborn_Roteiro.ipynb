## Criação de Subplots

````python
# Criar um array de valores x
x = np.linspace(0, 2 * np.pi, 100)
````

````python
# Criar figuras e subplots em uma grade 2x2
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plotar em cada subplot
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('Sen(x)')

axs[0, 1].plot(x, np.cos(x), 'tab:orange')
axs[0, 1].set_title('Cos(x)')

axs[1, 0].plot(x, np.tan(x), 'tab:green')
axs[1, 0].set_title('Tan(x)')

axs[1, 1].plot(x, np.exp(x), 'tab:red')
axs[1, 1].set_title('Exp(x)')

# Ajustar layout para evitar sobreposição de rótulos
plt.tight_layout()
plt.show()
````

## Criação de Gráficos a partir de Dicionários

````python
# Criar um DataFrame de exemplo
data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [10, 15, 7, 12, 20]
}
dff = pd.DataFrame(data)
````

````python
# Gráfico de linha
dff.plot(x='X', y='Y', kind='line')
plt.title('Gráfico de Linha')
plt.show()
````

````python
# Gráfico de barra
dff.plot(x='X', y='Y', kind='bar')
plt.title('Gráfico de Barra')
plt.show()
````

````python
# Gráfico de dispersão
dff.plot(x='X', y='Y', kind='scatter')
plt.title('Gráfico de Dispersão')
plt.show()
````

````python
# Gráfico de área
dff.plot(x='X', y='Y', kind='area')
plt.title('Gráfico de Área')
plt.show()
````

## Integração do Pandas com Matplotlib

Podemos definir o tipo de gráfico que desejamos trabalhar de duas formas. A primeira delas é passando o argumento kind dentro da função plot(), assim:

```df.plot(x, y, kind='bar')```

Outra forma de ser feito é utilizando a função específica para o gráfico desejado. Nesse exemplo em específico, seria:

 ```plt.bar(x,y)```

 ### Algumas dicas

- ```bar``` ou ```barh``` para bar plots

- ```hist``` para histograma

- ```box``` para boxplot

- ```kde``` ou ```density``` para density plots

````python
sns.histplot(data=titanic, x="Age", kde=True)
````

````python
sns.histplot(data=titanic, x="Age", hue="Survived", kde=True)
````

## Criação de Subplots

````python
# Carregue o conjunto de dados
data = sns.load_dataset("iris")

# Crie uma figura com subplots usando o Matplotlib
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 2 linhas, 2 colunas de subplots

# Crie gráficos individuais para cada subplot usando o Seaborn
sns.scatterplot(x="sepal_length", y="sepal_width", data=data, ax=axes[0, 0])
axes[0, 0].set_title("Comprimento vs Largura da Sépala")

sns.scatterplot(x="petal_length", y="petal_width", data=data, ax=axes[0, 1])
axes[0, 1].set_title("Comprimento vs Largura da Pétala")

sns.boxplot(x="species", y="sepal_length", data=data, ax=axes[1, 0])
axes[1, 0].set_title("Comprimento da Sépala por Espécie")

sns.boxplot(x="species", y="petal_length", data=data, ax=axes[1, 1])
axes[1, 1].set_title("Comprimento da Pétala por Espécie")

# Ajuste o layout dos subplots
plt.tight_layout()

# Exiba os subplots
plt.show()
````
